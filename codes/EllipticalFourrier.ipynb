```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def compute_elliptical_fourier_descriptors(contour, num_descriptors):
    contour_complex = contour[:, 0, :].astype(complex)
    
    fourier_descriptors = np.fft.fft(contour_complex, axis=0)
    
    descriptors = []
    for harmonic in range(1, num_descriptors + 1):
        real_part = fourier_descriptors[harmonic].real
        imag_part = fourier_descriptors[harmonic].imag
        amplitude = np.sqrt(real_part ** 2 + imag_part ** 2)
        phase = np.arctan2(imag_part, real_part)
        descriptors.append((amplitude, phase))
    
    return descriptors

# Example usage
image = cv2.imread('/home/stud1/Desktop/SURGE/Rice_Image_Dataset/Arborio/Arborio (1).jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
contour = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0][0], 0.01, True)

num_descriptors = 10
x = contour[:, 0, 0]
y = contour[:, 0, 1]


descriptors = compute_elliptical_fourier_descriptors(contour, num_descriptors)
reconstructed_contour = np.zeros_like(contour)
for t in np.linspace(0, 2 * np.pi, 100):
    contour_point = np.zeros((1, 2), dtype=np.float64)
    for n in range(len(descriptors)):
        contour_point[0, 0] += descriptors[n][0] * np.cos((n + 1) * t) + descriptors[n][2] * np.sin((n + 1) * t)
        contour_point[0, 1] += descriptors[n][1] * np.cos((n + 1) * t) + descriptors[n][3] * np.sin((n + 1) * t)
    reconstructed_contour = np.vstack((reconstructed_contour, contour_point))

plt.figure()
plt.plot(x, y, 'b', label='Original Contour')
plt.plot(reconstructed_contour[:, 0, 0], reconstructed_contour[:, 0, 1], 'r', label='Reconstructed Contour')
plt.legend()
plt.show()
# Plotting the amplitudes
"""amplitudes = [desc[0] for desc in descriptors]
plt.plot(range(1, num_descriptors + 1), amplitudes)
plt.xlabel('Harmonic')
plt.ylabel('Amplitude')
plt.title('Elliptical Fourier Descriptors')
plt.show()
"""
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def elliptical_fourier_descriptors(contour, num_coefficients):
    contour_reshaped = contour.reshape(-1, 2)
    contour_complex = contour_reshaped[:, 0] + 1j * contour_reshaped[:, 1]
    coefficients = np.fft.fft(contour_complex)
    descriptors = coefficients[:num_coefficients]
    return descriptors

def reconstruct_contour(descriptors, num_points):
    t = np.linspace(0, 2*np.pi, num_points)
    reconstructed_contour = np.zeros((num_points, 2), dtype=float)
    for i, point in enumerate(t):
        contour_point = np.zeros((1, 2), dtype=float)
        for n in range(len(descriptors)):
            amplitude = descriptors[n].real
            phase = descriptors[n].imag
            contour_point[0, 0] += amplitude * np.cos((n + 1) * point + phase)
            contour_point[0, 1] += amplitude * np.sin((n + 1) * point + phase)
        reconstructed_contour[i] = contour_point
    return reconstructed_contour

# Example usage
image = cv2.imread('/home/stud1/Desktop/SURGE/Rice_Image_Dataset/Arborio/Arborio (15000).jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
contour = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0][0]
num_coefficients = 10
num_points =5000

descriptors = elliptical_fourier_descriptors(contour, num_coefficients)
contour_reconstructed = reconstruct_contour(descriptors, num_points)

# Plotting the original and reconstructed contours
plt.figure()
#plt.plot(contour[:, 0, 0], contour[:, 0, 1], 'b', label='Original Contour')
plt.plot(contour_reconstructed[:, 0], contour_reconstructed[:, 1], 'r', label='Reconstructed Contour')
plt.gca().invert_yaxis()
plt.axis('equal')
plt.title('Original and Reconstructed Contours')
plt.legend()
plt.show()
```
